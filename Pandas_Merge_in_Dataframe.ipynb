```python
import pandas as pd

data1=[
    [1,'bala',25,150000],
    [2,'kumar',26,175000],
    [6,'balakumar',27,200000]
]

data2={
      "id":[1,2,3,4,5],
      "salary_increase":[10,20,30,40,50]
    }

schema1=['id','name','age','salary']

df1=pd.DataFrame(data=data1,columns=schema1)
df2=pd.DataFrame(data2)
```

```python
df3=pd.merge(df1,df2)
df3
```

```python
df3=pd.merge(df1,df2,how='outer',indicator=True)
df3
```

```python
data1={
    "id":[1,2,3,5,4],
    "name":['bala','kumar','balakumar','kumarbala','bkumar'],
    "age":[10,20,25,26,27]
}

data2={
    "id":[1,2,5,6,7],
    "name":['bala','kumar','balakumar','kumarbala','bkumar'],
    "age":[10,20,25,26,27]
}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df2
```

```python
df3=pd.concat([df1,df2],ignore_index=False,axis=0,keys=["id","name"])
df3
```

```python
df3=pd.merge(df1,df2,how='inner',on='name',suffixes=['_1','_2'])
df3
```
