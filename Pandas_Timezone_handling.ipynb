```python
import pandas as pd

df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pandas/msft.csv',header=1,index_col='Date Time',parse_dates=True)
df
```

```python
df.index
```

```python
df.index=df.index.tz_localize('Asia/Calcutta')
df.index
```

```python
df.index=df.index.tz_convert('Etc/GMT0')
df.index
```

```python
rng=pd.date_range(start='2017-08-17',periods=10,freq='H',tz='Africa/Algiers')
rng
```

```python
s=pd.Series(range(10),rng)
s
```

```python
s.index
```

```python
s=s.tz_convert('America/Argentina/Ushuaia')
s.index
```
