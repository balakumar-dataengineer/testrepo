```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pandas/demand_data_pandas.csv')
df
```

```python
df.replace(['Na','Not available'],np.nan)
```

```python
df.replace(
    {"Plant":[None],
     "Item":'Not available',
     "Date":'Na',
     "Demand":['Not available']
    },'0'
)
```

```python
df.replace({
    "Plant":{np.nan:'E80'},
    "Item":{np.nan:'Item_D',
            'Not available':'Item_D'},
    "Date":{np.nan:'08-03-2024',
            'Na':'08-03-2024'},
    "Demand":{'Not available':0}
})
```

```python
df.replace('[A-Za-z]','',regex=True)
```

```python
df.replace({
    "Date":'^[A-Za-z]+',
    "Demand":'^[A-Za-z]+$'
},'test',regex=True)
```
