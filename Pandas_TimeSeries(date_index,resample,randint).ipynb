```python
import pandas as pd

df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pandas/aapl.csv')
df
```

```python
rng=pd.date_range(start='2016-10-01',periods=251,freq='H')
rng
```

```python
df.set_index(rng,inplace=True)
df.sort_index(inplace=True)
```

```python
df.Close.mean()
```

```python
df.resample('W').mean()
```

```python
%matplotlib inline
df.Open.resample('QE').mean().plot(xlabel='Months',ylabel='price',kind='bar')
```

```python
df.Open.asfreq('D').plot()
```

```python
df.asfreq('D',method='bfill').head(10)
```
