```python
import pandas as pd

df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Pandas/msft.csv',header=1,parse_dates=True,index_col='Date Time')
df
```

```python
df['Price'].shift(2)
```

```python
df.shift(-4)
```

```python
df['previous_day_price']=df.shift(1)
df
```

```python
df['price difference']=df['Price']-df['previous_day_price']
df
```

```python
df['2 days % return']=(df['Price'] -df['Price'].shift(2))*100/df['Price'].shift(2)
df
```
