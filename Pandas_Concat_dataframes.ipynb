```python
import pandas as pd

emp_age={
    "name":['bala','kumar','balakumar'],
    "age":[25,26,27]
}

emp_sal={
    "name":['bala','balakumar','kumar'],
    "salary":[150000,180000,200000]
}

df1=pd.DataFrame(emp_age,index=[0,1,2])
df2=pd.DataFrame(emp_sal,index=[0,2,1])

print(df1.head())
df2.head()
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3=pd.concat([df1,df2],keys=['emp1','emp2'])
df3
```

```python
df3.loc['emp2']
```

```python
s=pd.Series(['1','2','3'],name='index')
s
```

```python
pd.concat([df1,df2,s],axis=1)
```
