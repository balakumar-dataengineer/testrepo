```python
import pandas as pd
data={
    'Country': ['Los Angeles', 'Los Angeles', 'Los Angeles',
               'Harris', 'Harris', 'Harris',
               'Cook', 'Cook', 'Cook'],

    'Date': ['2025-04-06', '2025-04-07', '2025-04-08',
             '2025-04-06', '2025-04-07', '2025-04-08',
             '2025-04-06', '2025-04-07', '2025-04-08'],

    'Condition': ['Sunny', 'Mostly sunny', 'Mostly sunny',
                  'Partly cloudy', 'Mostly sunny', 'Sunny',
                  'Mostly sunny', 'Cloudy', 'Partly sunny'],

    'High': [81, 76, 78,
                  59, 68, 74,
                  46, 40, 38],

    'Low': [52, 55, 55,
                 47, 48, 53,
                 36, 27, 33]
}

df=pd.DataFrame(data)
df
```

```python
df.pivot(index='Country',columns='Date',values='Condition')
```

```python
data = {
    'Country': ['CountyA', 'CountyA', 'CountyA', 'CountyA', 'CountyA', 'CountyA',
               'CountyB', 'CountyB', 'CountyB', 'CountyB', 'CountyB', 'CountyB',
               'CountyC', 'CountyC', 'CountyC', 'CountyC', 'CountyC', 'CountyC'],
    'Date': ['2025-04-01', '2025-04-01', '2025-04-02', '2025-04-02', '2025-04-03', '2025-04-03'] * 3,
    'TimeOfDay': ['Morning', 'Evening']*9,
    'Temperature': [18, 24, 19, 25, 17, 22,
                         20, 27, 21, 28, 19, 25,
                         16, 21, 17, 23, 15, 20],
    'Humidity': [80, 60, 78, 58, 82, 65,
                     75, 55, 73, 52, 77, 59,
                     85, 66, 83, 63, 88, 70]
}

df=pd.DataFrame(data)
df
```

```python
df.pivot_table(index='Country',columns='TimeOfDay',values=['Temperature','Humidity'],margins=True,aggfunc=lambda x: x.sum()/2) ## use both custom function and aggregate functions
```

```python
df.Date=pd.to_datetime(df.Date)
type(df.Date[0])
```

```python
df.pivot_table(index=pd.Grouper(freq='W',key='Date'),columns='TimeOfDay',values=['Humidity','Temperature'],aggfunc='mean')
```
